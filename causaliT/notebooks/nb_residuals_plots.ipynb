```python
from os.path import dirname, abspath
import sys
import matplotlib.pyplot as plt
import seaborn as sns


root_path = dirname(abspath("./"))
import sys
sys.path.append(root_path)
from causaliT.evaluation.predict import *
from notebooks.lib import *

# plotting standard settings
plt.rcParams['figure.dpi'] = 100 #360 # standard is 360 but for set to 100 for practical visualization on the notebook
plt.rcParams['axes.labelsize'] = 18
plt.rcParams['axes.titlesize'] = 18
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16
plt.rcParams['legend.fontsize'] = 14
plt.rcParams['figure.figsize'] = (6, 6)
plt.rcParams['lines.linewidth'] = 1.5
```

```python
df_1 = predict_with_error_analysis("../experiments/report/euler/stage_SoftMax_hard_wrong_underp_scm6_54694937")
df_1["experiment"] = "wrong DAG"
df_1["capacity"] = "low"

df_2 = predict_with_error_analysis("../experiments/report/euler/stage_SoftMax_hard_underp_scm6_54694944")
df_2["experiment"] = "true DAG"
df_2["capacity"] = "low"

df_3 = predict_with_error_analysis("../experiments/report/stage_SoftMax_hard_wrong_scm6")
df_3["experiment"] = "wrong DAG"
df_3["capacity"] = "high"

df_4 = predict_with_error_analysis("../experiments/report/euler/stage_SoftMax_hard_scm6_54603903")
df_4["experiment"] = "true DAG"
df_4["capacity"] = "high"

df = pd.concat([df_1, df_2, df_3, df_4], axis=0)
```

```python
df["mean_error_y"] = (df["abs_err_y_0"] + df["abs_err_y_1"]) / 2
df["mean_error_x"] = (df["abs_err_x_0"] + df["abs_err_x_1"]) / 2
df["mean_error"] = (df["mean_error_y"] + df["mean_error_x"]) / 2
```

```python
kfold_list =["k_0", "k_1", "k_2", "k_3", "k_4"]
svar_list = ["s1_value", "s2_value", "s3_value"]
experiment_list = ["true DAG", "wrong DAG"]
capacity_list = []
```

```python
hsic_results = []
for (capacity, experiment, kfold), group in df.groupby(["capacity", "experiment", "kfold"]):
    for s_var in svar_list:
        hsic = compute_hsic(group[s_var], group["mean_error"])
        hsic_results.append({
            "capacity": capacity,
            "experiment": experiment, 
            "kfold": kfold,
            "s_var": s_var,
            "hsic": hsic
        })
df_hsic = pd.DataFrame(hsic_results)
```

```python
df_tab_latex = pd.DataFrame()
df_tab_raw = df_hsic.groupby(["capacity", "experiment", "s_var"]).agg({"hsic":[np.mean, np.std]}).reset_index()
df_tab_latex["Regime"] = df_tab_raw[("capacity","")].map({"high": "over-par.", "low": "under-par."})
df_tab_latex["Experiment"] = df_tab_raw[("experiment","")]
df_tab_latex[r"$S$ variable"] = df_tab_raw[("s_var","")].map({"s1_value": r"$S_1$", "s2_value": r"$S_2$", "s3_value": r"$S_3$"})

scale = 1e5
df_tab_raw["mean_scaled"] = df_tab_raw[("hsic","mean")] * scale
df_tab_raw["std_scaled"] = df_tab_raw[("hsic","std")] * scale

df_tab_latex["HSIC"] = df_tab_raw.apply(
    lambda row: f"({row[('mean_scaled',"")]:.2f} ± {row[('std_scaled', "")]:.2f})E-5",
    axis=1
)

print(df_tab_latex.set_index(["Regime", "Experiment", r"$S$ variable"]).to_latex())
```

```python
for svar in svar_list:
    for kfold in kfold_list:
        plt.figure()
        sns.histplot(data=df.set_index(["capacity","kfold"]).loc["high"].loc[kfold], x=svar, y="mean_error", hue="experiment", multiple="stack", bins=100)
        plt.xlabel("mean residual")
        plt.xlabel(f"Normalized {svar}")
        plt.ylabel(r"Mean Absolute Residual")
        plt.title(f"Cross-Validation Fold {kfold}")
```
